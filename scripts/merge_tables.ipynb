```python
import pandas as pd

# Load the CSV files to inspect their headers
features_df = pd.read_csv('datasets-post-cleaning/features.csv')
gg_df = pd.read_csv('datasets-post-cleaning/gg_df_cleaned.csv')
oscars_df = pd.read_csv('datasets-post-cleaning/oscars_df_cleaned.csv')
sag_df = pd.read_csv('datasets-post-cleaning/sag_df_cleaned.csv')
```

```python
features_df.set_index(features_df.columns[1], inplace=True)
features_df.dropna(subset=['id'], inplace=True)
```

```python
features_df
```

```python
features_df = features_df.fillna(features_df.median())
features_df.reset_index(drop=False, inplace=True)
features_df
```

```python
# Rename the 'id' column in features_df to 'imdb_id' for consistency in merging

features_df['id'] = features_df['id'].astype(str)
features_df.rename(columns={'id': 'imdb_id'}, inplace=True)

# Convert 'imdb_id' in all dataframes to the same type for successful merge
features_df['imdb_id'] = features_df['imdb_id'].astype(str)
gg_df['imdb_id'] = gg_df['imdb_id'].astype(str)
oscars_df['imdb_id'] = oscars_df['imdb_id'].astype(str)
sag_df['imdb_id'] = sag_df['imdb_id'].astype(str)

features_df['imdb_id'] = features_df['imdb_id'].apply(lambda x: str(x).split('.')[0])
gg_df['imdb_id'] = gg_df['imdb_id'].apply(lambda x: str(x).split('.')[0])
oscars_df['imdb_id'] = oscars_df['imdb_id'].apply(lambda x: str(x).split('.')[0])
sag_df['imdb_id'] = sag_df['imdb_id'].apply(lambda x: str(x).split('.')[0])
```

```python
# Merge the dataframes on 'imdb_id'
merged_df = pd.merge(features_df, gg_df, on='imdb_id', how='left')
```

```python
merged_df = merged_df.fillna(0)
merged_df
```

```python
merged_df = pd.merge(merged_df, sag_df, on='imdb_id', how='left')
merged_df = merged_df.fillna(0)
merged_df
```

```python
merged_df = merged_df.drop(["film_x", "year_film", "film_y"], axis = 1)
```

```python
merged_df = pd.merge(merged_df, oscars_df, on='imdb_id', how='left')
merged_df = merged_df.fillna(0)
```

```python
merged_df
```

```python
merged_df = merged_df.drop(["year_x", "film", "year_y", "film.1"], axis = 1)
```

```python
merged_df.to_csv("merged_all_cleaned.csv", index=False)
```
